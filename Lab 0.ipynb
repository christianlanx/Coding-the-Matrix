```python
#0.5.30 Define a one-line procedure dict2list(dct, keylist) with the following spec:
# input: dictionary dict, list keylist consisting of the keys of dct
# output: list L such that L[i] = dct[keylist[i]] for i = 0, 1, 2, ..., len(keylist) -1
#output = ['B','C','A']
def dict2list(dct, keylist): return [dct[x] for x in keylist]
```

```python
A = {'a':'A','b':'B','c':'C'}
B=['b','c','a']
def dict2list(a,b): return [a[x] for x in b]
dict2list(A,B)
```

```python
# 0.5.31 Define a one-line procedure list2dict(L, keylist) specifies as follows:
# input: list L, list keylist of immutable items
# output: dictionary that maps keylist[i] to L[i] for i = 0, 1, 2,...,len(L)-1
# example: input L=['A','B','C'] and keylist=['a','b','c']
#       output{'a':'A','b':'B','c':'C'}
def list2dict(listA,listB): return{a:b for (a,b) in zip(listA,listB)}
L = ['A','B','C']
keylist = ['a','b','c']
list2dict(keylist,L)
```
